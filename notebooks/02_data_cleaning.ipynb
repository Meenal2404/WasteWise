```python
import pandas as pd
import numpy as np

# Load raw data
df = pd.read_csv('../data/raw/wastewise_raw.csv')

print("Data loaded successfully!")
print("Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Add useful columns
df['month'] = df['date'].dt.month_name()
df['day_of_week'] = df['date'].dt.day_name()
df['month_num'] = df['date'].dt.month

# Standardize text columns
df['area'] = df['area'].str.strip()
df['waste_type'] = df['waste_type'].str.strip()
df['collection_done'] = df['collection_done'].str.strip()

print("Data cleaned successfully!")
print("\nUpdated Data types:")
print(df.dtypes)
print("\nNew columns added:")
print(df[['date', 'month', 'day_of_week']].head())

# Save cleaned data
df.to_csv('../data/cleaned/wastewise_cleaned.csv', index=False)
print("\nCleaned data saved!")
```
