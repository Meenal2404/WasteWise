```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Settings
random.seed(42)
np.random.seed(42)

areas = ['Arera Colony', 'Shahpura', 'TT Nagar', 
         'Kolar Road', 'Bairagarh', 'Govindpura']
waste_types = ['Wet', 'Dry', 'Medical', 'Construction']
start_date = datetime(2025, 1, 1)

# Generate 6 months data
rows = []
for i in range(180):
    date = start_date + timedelta(days=i)
    for area in areas:
        rows.append({
            'date': date.strftime('%Y-%m-%d'),
            'area': area,
            'waste_type': random.choice(waste_types),
            'quantity_kg': round(random.uniform(50, 500), 2),
            'collection_done': random.choice(['Yes', 'No']),
            'complaints': random.randint(0, 10)
        })

df = pd.DataFrame(rows)
df.to_csv('../data/raw/wastewise_raw.csv', index=False)
print("Data generated successfully!")
print("Total rows:", len(df))
print(df.head())
```
